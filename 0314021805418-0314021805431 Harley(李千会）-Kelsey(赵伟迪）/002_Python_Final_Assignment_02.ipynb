````python
# Solution:

import random
import string
import re
import sys
def guess():
    maps = ["""____ 
  |    | 
  |    O 
  |   /|  """,
 """  ____ 
  |    | 
  |    O 
  |   /|\
""",
"""
   ____ 
  |    | 
  |    | 
  |    O 
  |   /|\ 
  |   / \ 
__|__ 
""" ]
    with open('./sowpods.txt','r', encoding="utf8") as f:
        words = [i.replace("\n","") for i in f.readlines(10000)]
    word = random.choice(words)
    count = 7
    word_vec = [i.upper() for i in list('abcdefghijklmnopqrstuvwxyz')]
    str_init = ['_' for i in range(len(word))]
    wrong = []
    while True:
        guess = input("What is your guess?")
        if guess.upper() not in word_vec:
            guess = input("Please chose just a letter: ")
        if guess.upper() in wrong or (guess.upper() in ''.join(str_init)):
            print(f"You have already guessed {guess}!")
            print(''.join(str_init))
            guess = input("What is your guess?")
        if (guess.upper() in word):
            position = [i.start() for i in re.finditer(guess.upper(), word)]

            for i in position:
                str_init[i] = guess.upper()
            print(''.join(str_init))

        else:
            print(f"{guess} is not in this word!")
            print(random.choice(maps))
            wrong.append(guess.upper())
            count -= 1
            print(f"you have {count} guesses left!")
        if len(wrong) >= 1:
            print(f"your previous wrong guesses: {list(set(wrong))}")
            print(''.join(str_init))
        if ('_' not in ''.join(str_init)):
            print("You win")
            break
        if count == 0 and ('_' in ''.join(str_init)):
            print("You have 0 guesses left!")
            print("You lose!")
            print(f"your previous wrong guesses: {list(set(wrong))}")
            print(f"The word was {list(word)}")
            choose = input("Would you like to play again?[y|n]:")
            if choose == 'n':
                return False
            elif choose == 'y':
                return True

            break
    return True
if __name__ == '__main__':
    while True:
        bool = guess()
        if not bool:
            sys.exit(0)
            break
````
